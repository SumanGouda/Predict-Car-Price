```python
import pandas as pd
import json

def convert_json_to_csv(json_file_path, csv_output_path):
    data = []
    
    try:
        with open(json_file_path, 'r', encoding='utf-8') as f:
            for line in f:
                # Skip empty lines
                if line.strip():
                    # Load each line individually
                    data.append(json.loads(line))
        
        # 2. Convert the list of objects to a DataFrame
        df = pd.DataFrame(data)
        
        # 3. Save to CSV (following your replacement rule later)
        df.to_csv(csv_output_path, index=False, encoding='utf-8-sig')
        
        print(f"✅ Success! Merged multiple JSON objects.")
        print(f"Total Rows: {len(df)} | Total Columns: {len(df.columns)}")

    except Exception as e:
        print(f"❌ Error: {e}")

# Usage
convert_json_to_csv('car_dataset_delhi.json', 'cleaned_car_data_delhi.csv')
```
