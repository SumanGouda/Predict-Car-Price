```python
import pandas as pd
import json

def convert_json_to_csv(json_file_path, csv_output_path):
    # 1. Load the JSON data
    # This handles standard JSON lists of objects
    with open(json_file_path, 'r') as f:
        data = json.load(f)
    
    # 2. Convert to DataFrame
    df = pd.DataFrame(data)
    
    # 3. Save to CSV
    # index=False prevents pandas from adding an extra 'Unnamed: 0' column
    df.to_csv(csv_output_path, index=False)
    
    print(f"Success! '{json_file_path}' converted to '{csv_output_path}'")
    print(f"Total Rows: {len(df)} | Total Columns: {len(df.columns)}")

# Usage
convert_json_to_csv('your_data.json', 'cleaned_car_data.csv')
```
