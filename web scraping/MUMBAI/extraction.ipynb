## Extracting the car links from the main webpage

```python
browser = Browser('chrome')

base_url = "https://www.cardekho.com/used-cars+in+mumbai"
```

```python
# 1. Open a text file in 'a' (append) mode
with open("car_links.txt", "a") as f:
    target_cars = 5000
    cars_collected = 0
    total_pages = 250 

    for page_num in range(1, total_pages + 1):
        if page_num == 1:
            url = "https://www.cardekho.com/used-cars+in+mumbai"
        else:
            url = f"https://www.cardekho.com/used-cars+in+mumbai/page-{page_num}"
        
        browser.visit(url)
        time.sleep(2) 
        
        browser.execute_script("window.scrollTo(0, 1000);")
        time.sleep(1)

        current_soup = soup(browser.html, 'html.parser')
        
        page_links_found = 0
        for link in current_soup.find_all('a', href=True):
            href = link['href']
            
            if 'used-car-details' in href:
                # Build the full URL
                full_url = f"https://www.cardekho.com{href}" if href.startswith('/') else href
                
                # 2. Write the link to the text file immediately
                f.write(full_url + "\n")
                
                cars_collected += 1
                page_links_found += 1

        print(f"Page {page_num}: Saved {page_links_found} links. Total: {cars_collected}")

        # Stop if target reached
        if cars_collected >= target_cars:
            print("Target 5000 reached!")
            break

print("All links are now saved in car_links.txt")
```

## Main Extraction 

```python
def get_browser():
    return Browser('chrome')

# 1. Load your links
with open("car_links.txt", "r") as f:
    all_links = [line.strip() for line in f.readlines()]

browser = get_browser()
output_file = "car_dataset_mumbai.json"

# 2. Loop through each link
for index, link in enumerate(all_links):
    try:
        print(f"Scraping {index+1}/{len(all_links)}: {link}")
        
        # --- CRITICAL FIX: Visit the link first ---
        browser.visit(link)
        
        # Human-like delay and scroll
        time.sleep(random.uniform(2, 4))
        browser.execute_script("window.scrollTo(0, 800);")
        time.sleep(1)

        # 3. Expansion Logic
        try:
            # We use a broader search for the button text
            view_all_spec_btn = browser.find_by_text('View all Specifications')
            if view_all_spec_btn:
                browser.execute_script("arguments[0].click();", view_all_spec_btn.first._element)
                print("Successfully expanded specs.")
                time.sleep(2) # Wait for table to render
        except Exception as e:
            print(f"Expansion skipped (might already be open): {e}")

        # 4. Scrape with BeautifulSoup
        page_soup = soup(browser.html, 'html.parser')
        car_data = {"url": link}
        
        # Extract labels and values
        spec_items = page_soup.find_all('li', class_='gsc_col-xs-12')
        
        for item in spec_items:
            label_tag = item.find('div', class_='label')
            value_tag = item.find('span', class_='value-text')
            if label_tag and value_tag:
                label = label_tag.get_text(strip=True)
                value = value_tag.get_text(strip=True)
                car_data[label] = value

        # --- SAFETY CHECK: Only save if we got more than just the URL ---
        if len(car_data) > 1:
            with open(output_file, "a") as out:
                out.write(json.dumps(car_data) + "\n")
            print(f"Saved {len(car_data)-1} specifications.")
        else:
            print("WARNING: No data found for this link. Check selectors.")

    except Exception as e:
        print(f"Serious error at {link}: {e}")
        browser.quit()
        browser = get_browser()
        time.sleep(5) # Cooldown before restart
        continue

browser.quit()
```

### Extracting the Price Columns

```python
def get_browser():
    return Browser('chrome')

# 1. Load the original links and existing car data
with open("car_links.txt", "r") as f:
    all_links = [line.strip() for line in f.readlines()]

existing_cars = []
with open("car_dataset_mumbai.json", "r") as f:
    for line in f:
        existing_cars.append(json.loads(line))

browser = get_browser()

# 2. Iterate through both in perfect sync
# We use zip to ensure we are looking at the exact same car index
for index, (link, car_data) in enumerate(zip(all_links, existing_cars)):
    try:
        print(f"Updating Price for {index+1}/{len(existing_cars)}: {link}")
        
        browser.visit(link)
        time.sleep(random.uniform(.1, .5))
        
        page_soup = soup(browser.html, 'html.parser')
        
        # --- The Price Extraction Logic ---
        price_div = page_soup.find('div', class_='vehiclePrice')
        if price_div:
            price_span = price_div.find('span')
            if price_span:
                # Add price to the existing dictionary
                car_data["Price"] = price_span.get_text(strip=True)
                print(f"Success: {car_data['Price']}")
        
        # 3. Save the updated line immediately to a NEW file (safer)
        with open("car_dataset_mumbai_updated.json", "a") as out:
            out.write(json.dumps(car_data) + "\n")

    except Exception as e:
        print(f"Error at {link}: {e}")
        # Save without price if extraction fails to maintain order
        with open("car_dataset_mumbai_updated.json", "a") as out:
            out.write(json.dumps(car_data) + "\n")
        continue

browser.quit()
```

```python
import json
import csv

def json_to_csv(json_file, csv_file):
    data = []
    all_keys = set()

    # 1. Read JSON Lines and collect all possible field names
    try:
        with open(json_file, 'r') as f:
            for line in f:
                if line.strip():
                    item = json.loads(line)
                    data.append(item)
                    all_keys.update(item.keys())
    except FileNotFoundError:
        print(f"Error: {json_file} not found.")
        return

    if data:
        # Convert set to list for a consistent header order
        header = sorted(list(all_keys))
        
        # 2. Write to CSV using DictWriter to handle missing values automatically
        with open(csv_file, 'w', newline='', encoding='utf-8') as f:
            writer = csv.DictWriter(f, fieldnames=header)
            writer.writeheader()
            writer.writerows(data)
        
        print(f"Successfully converted {len(data)} rows to {csv_file}")
    else:
        print("No data found in the JSON file.")

# Usage
json_to_csv('car_dataset_mumbai.json', 'car_dataset_mumbai.csv')
```

```python
import pandas as pd

# 1. Load the JSON Lines file
df = pd.read_json('car_dataset_mumbai.json', lines=True)

# 2. Clean up column names (CarDekho often has spaces or special characters)
# This makes the CSV headers look professional
df.columns = [col.strip() for col in df.columns]

# 3. Save to CSV
# index=False prevents an extra "0, 1, 2..." column from appearing
df.to_csv('car_dataset_mumbai.csv', index=False, encoding='utf-8-sig')

print("Your organized CSV is ready!")
```
